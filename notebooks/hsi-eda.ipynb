```python
from pathlib import Path
from collections import Counter

root = Path("/home/brwsx/hsi-compression/pipelines/pull-dataset/hyspectnet-11k")

files = [p for p in root.rglob("*") if p.is_file()]
print("Total files:", len(files))

exts = Counter(p.suffix.lower() for p in files)
print("Extensions:")
for ext, count in exts.most_common():
    print(f"{ext}: {count}")

print("\nFirst 30 files:")
for p in files[:30]:
    print(p)
```

Dataset does not have one simple file per sample. Instead, each patch directory seems to contain:
- one main hyperspectral cube: *-SPECTRAL_IMAGE.TIF
- several quality/mask files (cloud/snow/haze/cirrus/class flags):
    - *-QL_PIXELMASK.TIF
    - *-QL_VNIR.TIF
    - *-QL_SWIR.TIF
- one thumbnail: *-THUMBNAIL.jpg

So for compression, my primary input is almost certainly:
SPECTRAL_IMAGE.TIF

```python
from pathlib import Path
import tifffile as tiff
import numpy as np

root = Path("/home/brwsx/hsi-compression/pipelines/pull-dataset/hyspectnet-11k/hyspecnet-11k-full")
sample = next(root.rglob("*-SPECTRAL_IMAGE.TIF"))

x = tiff.imread(sample)

print("Sample path:", sample)
print("Shape:", x.shape)
print("Dtype:", x.dtype)
print("Min:", np.min(x))
print("Max:", np.max(x))
print("Mean:", np.mean(x))
print("Std:", np.std(x))
```

```python
import tifffile as tiff
from pathlib import Path

root = Path("/home/brwsx/hsi-compression/pipelines/pull-dataset/hyspectnet-11k/hyspecnet-11k-full")
sample = next(root.rglob("*-SPECTRAL_IMAGE.TIF"))

with tiff.TiffFile(sample) as tif:
    print("Number of pages:", len(tif.pages))
    page0 = tif.pages[0]
    print("Page 0 shape:", page0.shape)
    print("Page 0 dtype:", page0.dtype)
    print("Series count:", len(tif.series))
    for i, s in enumerate(tif.series):
        print(f"Series {i}: shape={s.shape}, axes={getattr(s, 'axes', 'unknown')}")
```
